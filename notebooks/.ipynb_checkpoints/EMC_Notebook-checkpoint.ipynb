## [R] Import packages
Import the necessary Python packages and the EMC utilities package

```python
# Importing packages.
import os
import sys
import getopt
import math
from datetime import datetime, timezone
from IPython.display import Markdown, display
from netCDF4 import Dataset
import xarray as xr
import matplotlib.pyplot as plt

# EMC's utility functions.
import emc_utils as utils
from emc_utils import get_var_name
```

## [R] Configuration
### Initialize code parameters and prepare for execution

User may modify the following variables for further customization:
* DELIMITER - The separator to use when storing data in GeoCSV file. 
* DATA_DIR - Directory under this Notebook's directory that would be used to read files from.
* OUT_DIR - Directory under this Notebook's directory that would be used to store files to.
* NETCDF_EXTENSION - The file extension for the netCDF files.

```python
# Initialize parameters.

# Delimiter for the GeoCSV file.
DELIMITER = '|'

# Directory under the script's directory to read netCDF files from. 
DATA_DIR = 'data'

# Directory under the script's directory to store image and data files. 
# Will create if it does not exist.
OUT_DIR = 'output'

# Default extension for the netCDF files.
NETCDF_EXTENSION = '.nc'

# ---------------------------------

# Container for the run parameters.
run_args = dict()

VERSION = "r0.0.0"
run_args['VERSION'] = VERSION

# The script's name and its version.
SCRIPT = sys.argv[0]
run_args['SCRIPT'] = SCRIPT

# Get the root _directory.
ROOT_DIR = os.getcwd()
run_args['ROOT_DIR'] = ROOT_DIR

# GeoCSV version to use.
GEOCSV_VERSION = 'GeoCSV2.0'
run_args['GEOCSV_VERSION'] = GEOCSV_VERSION

# Make sure the data directory is accessible.
DATA_DIR = os.path.join(ROOT_DIR, DATA_DIR)
if not os.path.isdir(DATA_DIR):
    raise ValueError(f"Could not access the data directory {DATA_DIR}"
                     f"\nStop! Please update the 'DATA_DIR' value")
run_args['DATA_DIR'] = DATA_DIR

# Make sure the output directory is accessible.
# Create if it does not exist.
try:
    OUT_DIR = utils.make_path(os.path.join(ROOT_DIR, OUT_DIR))   
except:
    raise ValueError(f"Could not access or create the data directory {OUT_DIR}"
                     f"\nStop! Please update the 'OUT_DIR' value")
    
run_args['DELIMITER'] = DELIMITER

# Some control paramaters.
VALID_MODES = {'depth': 'km', 'single': ''}
VALID_DIMENSIONS = [2, 3]
NETCDF_FILE_NAME = None
BASE_NAME = None
```

## [O] List model files under the data directory.

Run the following cell to see  a list of all available netCDF files under the data directory. Then select a file and assign it to the **MODEL_FILE** variable. If you already know the file name, you may skip the next cell.

```python
# Provide a list of candidate (.nc) files.
found_file = False
display(Markdown(f"\n\n## List of all available netCDF files\n### {DATA_DIR}:\n\n"))
for _file in os.listdir(DATA_DIR):
    if _file.lower().endswith(NETCDF_EXTENSION):
        print(f"\t{_file}")
        found_file = True

if not found_file:
    raise UserWarning(f"Stop! No netCDF file was found under the data '{DATA_DIR}' directory")
else:
    display(Markdown(f"\n\n<font color='red'>Select one of the above files and assign it to the **MODEL_FILE** variable in the cells below.</font>"))
    
```

## [R] Load the model
### Load a netCDF model file

In the following cell, assign the name of the netCDF model file you want to explore to the **MODEL_FILE** variable. 
The file must reside under the data directory assigned to the **DATA_DIR** variable in the previously. 

```python
# Model file name. Do not include path but must include extension (.nc).
# A None file name will list all available files (.nc).
MODEL_FILE = "KEA20.nc"

# Model file name is given, make sure it exists.
model_file = os.path.join(DATA_DIR, MODEL_FILE)
if os.path.isfile(model_file):
    display(Markdown(f"\n\n## Loading the model file:\n{model_file}\n"))
else:
    raise UserWarning(f"Could not find model file '{model_file}'"
                      "Stop! Please update the 'MODEL_FILE' and/or 'DATA_DIR' values")
 
# Base file name.
base_file_name = os.path.splitext(MODEL_FILE)[0]

# Create an output directory for this model.
try:
    OUT_DIR = utils.make_path(os.path.join(OUT_DIR, base_file_name))   
except:
    raise ValueError(f"Could not access or create the data directory {OUT_DIR}"
                     f"\nStop! Please update the 'OUT_DIR' value")
    
# All good with the model file. Read it. 
if MODEL_FILE is not None:
    model_data = Dataset(model_file)
    run_args['model_data'] = model_data
    run_args['MODEL_FILE'] = MODEL_FILE

    ds = xr.open_dataset(model_file)
    run_args['ds'] = ds

    # Tell the user that the model file is loaded.
    display(Markdown(f"\n\n## ...loaded\n\n"))
```

## [R] Model coordinates
Modify the variables in the next cell to define variables for model coordinates. To load a 2D model, set the **Z_VARIABLE** variable to **None** 

```python
# Model dimensions.
# Must match the netCDF file's x variable.
X_VARIABLE = 'longitude'  
run_args['X_VARIABLE'] = X_VARIABLE
x_values = model_data.variables[X_VARIABLE][:]
run_args['x_values'] = x_values

# Must match the netCDF file's y variable.
Y_VARIABLE = 'latitude' 
run_args['Y_VARIABLE'] = Y_VARIABLE
y_values = model_data.variables[Y_VARIABLE][:]
run_args['y_values'] = y_values

# must match the netCDF file's z variable. For the 2D models, set to None.
Z_VARIABLE = 'depth'  
run_args['Z_VARIABLE'] = Z_VARIABLE
z_values = model_data.variables[Z_VARIABLE][:]
run_args['z_values'] = z_values
```

## [R] Update/validate user inputs
The following cell updates (if needed) and validates the provided parameters before we continue.

```python
# Update/validate parameters based on user inputs. 

if X_VARIABLE is None or Y_VARIABLE is None:
    raise ValueError(f"The 'X_VARIABLE' or 'Y_VARIABLE' variables cannot be set to 'None'"
                     f"\nStop! Please update the 'X_VARIABLE' and/or 'Y_VARIABLE' values")

model_var_list = list(model_data.variables)
run_args['model_var_list'] = model_var_list

if X_VARIABLE not in model_var_list:
    raise UserWarning(f"' X_VARIABLE' set as '{X_VARIABLE}' is not a model variable"
                          "Stop! Please update the 'X_VARIABLE' variable")
if Y_VARIABLE not in model_var_list:
    raise UserWarning(f"' Y_VARIABLE' set as '{Y_VARIABLE}' is not a model variable"
                          "Stop! Please update the 'Y_VARIABLE' variable")
if Z_VARIABLE is None:
    NDIM = 2
else:
    NDIM = 3
    if Z_VARIABLE not in model_var_list:
        raise UserWarning(f"' Z_VARIABLE' set as '{Z_VARIABLE}' is not a model variable"
                              "Stop! Please update the 'Z_VARIABLE' variable")
run_args['NDIM'] = NDIM

ndim_label = f"{NDIM}D"
run_args['ndim_label'] = ndim_label
```

## [O] Plot horizontal (depth) slice(s) of the model.
Plot one or more horizontal slices of the model at given depth(s) for a defined area. 
To make the plots, in the next cell first assigned the desired model variable to 
the **VARIABLE** variable. Then define the range (min, max) for each of the coordinate variables.
A blank range (**[]**) will make plots of all values. The **Z_RANGE** has no effect for the 2D models. 
For the #D models, change the **Z_STEP** value to decimate (resample) the depth values (set **Z_STEP = 1** for all depths).
You can also assign a colormap to the plots. For a list of available colormaps, see https://matplotlib.org/stable/tutorials/colors/colormaps.html.

## [O] Plot vertical (cross section) slice(s) of a 3D model along the X-axis (longitude)
### 3D models only
Plot one or more vertical slices of a 3D model along X-axis for a defined Y and Z ranges. 
To make the plots, in the next cell first assigned the desired model variable to 
the **VARIABLE** variable. Then define the range (min, max) for each of the coordinate variables.
A blank range (**[]**) for a coordinate will make plots of all values along that coordinate. 
Change the **X_STEP** value to decimate (resample) the X values (set **X_STEP = 1** for all X values).
You can also assign a colormap to the plots. For a list of available colormaps, see https://matplotlib.org/stable/tutorials/colors/colormaps.html.

```python
# Assigne the desired variable.
VARIABLE = 'vsv'

# matplotlib colormap to use. For available colormaps see:
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
COLORMAP ='jet_r'

# Range of variables [minimum, maximum] to use. To output all, set to empty ([]). 
# Z values are used only for 3D models. Use Z_STEP to decimate (resample) the depth values (set Z_STEP = 1 for all depths).
X_RANGE = [] # [140,146]
Y_RANGE = [] # [20,40]
Z_RANGE = [40,170]
Z_STEP = 2

# Save plot and data flags. SAVE_TAG will be added to the file names.
SAVE_TAG = "test"
SAVE_PLOTS = True
SAVE_DATA = True

# ---------------------------------
x_min, x_max, y_min, y_max, z_min, z_max = utils.get_ranges(X_RANGE, x_values, Y_RANGE, y_values, Z_RANGE, z_values, NDIM)
    
if NDIM == 3:
    Z = list()  
    for _z in z_values:
        if z_min <= _z <= z_max:
            Z.append(_z)
    
dsv = ds[VARIABLE]
if NDIM == 3:
    for depth in Z[::Z_STEP]:
        zslice = dsv.where((dsv.longitude >= x_min) & (dsv.longitude <= x_max) &
                                (dsv.latitude >= y_min) & (dsv.latitude <= y_max) &
                                (dsv.depth == depth), drop=True)
        zslice.plot(cmap=COLORMAP)
        save_file = os.path.join(OUT_DIR, '_'.join([base_file_name, VARIABLE, f"Z{depth}km", f"{SAVE_TAG}"]))
        if SAVE_DATA:
            zdf = zslice.to_dataframe()
            zdf = zdf.to_csv(f"{save_file}.csv", sep=DELIMITER)
            
        if SAVE_PLOTS:
            plt.savefig(f"{save_file}.png")
        plt.show()
else:
    dsv2 = dsv.where((dsv.longitude >= x_min) & (dsv.longitude <= x_max) &
                                (dsv.latitude >= y_min) & (dsv.latitude <= y_max), drop=True)
    
    dsv2.plot(cmap=COLORMAP)
    save_file = os.path.join(OUT_DIR, '_'.join([base_file_name, VARIABLE, 'hslice', f"{SAVE_TAG}"]))

    if SAVE_DATA:
        df = dsv2.to_dataframe()
        df = df.to_csv(f"{save_file}.csv", sep=DELIMITER)
        
    if SAVE_PLOTS:
        plt.savefig(f"{save_file}.png")
    plt.show()
```

## [O] Plot vertical (cross section) slice(s) of a 3D model along the Y-axis (longitude)
### 3D models only
Plot one or more vertical slices of a 3D model along Y-axis for a defined X and Z ranges. 
To make the plots, in the next cell first assigned the desired model variable to 
the **VARIABLE** variable. Then define the range (min, max) for each of the coordinate variables.
A blank range (**[]**) for a coordinate will make plots of all values along that coordinate. 
Change the **Y_STEP** value to decimate (resample) the Y values (set **Y_STEP = 1** for all Y values).
You can also assign a colormap to the plots. For a list of available colormaps, see https://matplotlib.org/stable/tutorials/colors/colormaps.html.

```python
if NDIM != 3:
    raise UserWarning(f"Plot of vertical (cross section) slice(s) is nly available for 3D models")

# Assigne the desired variable.
VARIABLE = 'vsv'

# matplotlib colormap to use. For available colormaps see:
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
COLORMAP ='jet_r'

# Range of variables [minimum, maximum] to use. To output all, set to empty ([]). 
# Use the X_STEP to decimate (resample) the X values (set X_STEP = 1 for all X values).
X_RANGE = [] # [140,146]
X_STEP = 50

Y_RANGE = [] # [20,40]
Z_RANGE = [40,170]

# Save plot and data flags. SAVE_TAG will be added to the file names.
SAVE_TAG = "test"
SAVE_PLOTS = True
SAVE_DATA = True

# ---------------------------------
x_min, x_max, y_min, y_max, z_min, z_max = utils.get_ranges(X_RANGE, x_values, Y_RANGE, y_values, Z_RANGE, z_values, NDIM)
    
X = list()
for _x in x_values:
    if x_min <= _x <= x_max:
        X.append(_x)
    
dsv = ds[VARIABLE]
for _x in X[::X_STEP]:
    xsection = dsv.where((dsv.latitude >= y_min) & (dsv.latitude <= y_max) &
                            (dsv.depth >= z_min) & (dsv.depth <= z_max) &
                            (dsv.longitude == _x), drop=True)
    xsection.plot(x="latitude", y="depth", cmap=COLORMAP)

    save_file = os.path.join(OUT_DIR, '_'.join([base_file_name, VARIABLE, f"X{_x:0.1f}km", f"{SAVE_TAG}"]))
    if SAVE_DATA:
        xdf = xsection.to_dataframe()
        xdf = xdf.to_csv(f"{save_file}.csv", sep=DELIMITER)

    if SAVE_PLOTS:
        plt.savefig(f"{save_file}.png")
    plt.show()
```

## [O] Plot vertical depth profiles along X-axis (longitude)
### 3D models only
Plot one or more vertical profiles (change of parameter with depth) for a 3D model along X-axis for selected Y and Z ranges. 
To make the plots, in the next cell:
* set **VARIABLE** to the desired model variable
* define the range (min, max) for each of the coordinate variables ( a blank range (**[]**)  will use all values 
* use **X_STEP** &amp; **Y_STEP** to decimate (resample every {X|Y}_STEP samples) the {X|Y} values (**{X|Y}_STEP = 1** for all {X|Y} values)
* assign a colormap to the plots (for available colormaps see https://matplotlib.org/stable/tutorials/colors/colormaps.html)
* set **SAVE_PLOTS** &amp; **SAVE_DATA** to **True** if you want to save the generated plots and the corresponding data
* set **SAVE_TAG** to a string to be included in the plot and data file names

```python
if NDIM != 3:
    raise UserWarning(f"Plot of vertical (cross section) slice(s) is nly available for 3D models")

# Assigne the desired variable.
VARIABLE = 'vsv'

# Range of variables [minimum, maximum] to use. To output all, set to empty ([]). 
# Use the Y_STEP to decimate (resample) the Y values (set Y_STEP = 1 for all X values).
X_RANGE = []
X_STEP = 100

Y_RANGE = []
Y_STEP = 50

Z_RANGE = []

# A matplotlib colormap to use. For available colormaps see:
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
COLORMAP ='jet_r'

# Save plot and data flags. SAVE_TAG will be added to the file names.
SAVE_PLOTS = True
SAVE_DATA = True
SAVE_TAG = "test"

# ---------------------------------
x_min, x_max, y_min, y_max, z_min, z_max = utils.get_ranges(X_RANGE, x_values, Y_RANGE, y_values, Z_RANGE, z_values, NDIM)

X = utils.get_range_values(x_values, (x_min, x_max))
Y = utils.get_range_values(y_values, (y_min, y_max))
        
dsv = ds[VARIABLE]
for _x in X[::X_STEP]:
    profile = dsv.sel(longitude =_x, latitude= Y[::Y_STEP])
    
    profile.plot.line(y='depth', hue='latitude')
    plt.gca().invert_yaxis()
    
    save_file = os.path.join(OUT_DIR, '_'.join([base_file_name, VARIABLE, f"XP{_x:0.1f}deg", f"{SAVE_TAG}"]))
    if SAVE_DATA:
        xdf = xsection.to_dataframe()
        xdf = xdf.to_csv(f"{save_file}.csv", sep=DELIMITER)

    if SAVE_PLOTS:
        plt.savefig(f"{save_file}.png")
    plt.show()
```

## Output the model in GeoCSV format
To output the model or a selected portion of it in GeoCSV format, in the next cell set the **OUTPUT_GEOCSV** variable to **True**. You may also set range for variables to output

```python
# Output the GeoCSV model

# depth: One output file per depth.
# single: a single output file that contains all depths.
OUTPUT_MODE = 'single'
run_args['VIEW_HEADER'] = VIEW_HEADER

if OUTPUT_MODE not in list(VALID_MODES):
    raise UserWarning(f"Invalid 'OUTPUT_MODE' of '{OUTPUT_MODE}' set OUTPUT_MODE to one of {VALID_MODES} types."

                       "Stop! Please update the 'OUTPUT_MODE' variable")
if len(X_RANGE) not in (0, 2):
    raise UserWarning(f"Invalid 'X_RANGE' of '{X_RANGE}' must be either and empty range for all ([]) for sould have min and max values"
                       "Stop! Please update the 'X_RANGE' variable")

if len(Y_RANGE) not in (0, 2):
    raise UserWarning(f"Invalid 'Y_RANGE' of '{Y_RANGE}' must be either and empty range for all ([]) for sould have min and max values"
                       "Stop! Please update the 'Y_RANGE' variable")

if len(Z_RANGE) not in (0, 2):
    raise UserWarning(f"Invalid 'Z_RANGE' of '{Z_RANGE}' must be either and empty range for all ([]) for sould have min and max values"
                       "Stop! Please update the 'Z_RANGE' variable")

utils.make_model_geocsv(run_args, OUTPUT_MODE)
print("...done")
```
